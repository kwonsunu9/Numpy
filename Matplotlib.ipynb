```python
from polynomial import Polynomial
import numpy as np
import matplotlib.pyplot as plt
p = Polynomial(-0.8, 2.3, 0.5, 1, 0.2)
p_der = p.derivative
X = np.linspace(-2, 3, 50, endpoint=True)
F = p(X)
F_derivative = p_der(X)
fig, ax = plt.subplots()

ax.plot(X, F)
ax.plot(X, F_derivative)
ax.legend(["P", "derivation of P"], loc = "center")
plt.show()
```

```python
p = Polynomial(-0.8, 2.3, 0.5, 1, 0.2)
p_der = p.derivative
X = np.linspace(-2, 3, 50, endpoint=True)
F = p(X)
F_derivative = p_der(X)
fig, ax = plt.subplots()

ax.plot(X, F, label = "P")
ax.plot(X, F_derivative, label = "derivation of P")
ax.legend(loc = "upper right")
plt.show()
```

#### ANNOTATIONS
* matplotlib.pyplot.annotate(text, xy, xytext=None, xycoords='data', textcoords=None, arrowprops=None, annotation_clip=None, **kwargs)

```python
X = np.arange(-5, 5, 0.1)
p = Polynomial(1, 0, -12, 0)
p_der = p.derivative
F = p(X)
F_derivative = p_der(X)

fig, ax = plt.subplots()
ax.plot(X, F, label = f"$ {p}")
ax.plot(X, F_derivative, label = f"$ {p_der} $")
ax.annotate("local maximum", (-2, p(-2)))
ax.annotate("local minimum", (2, p(2)))
ax.legend(loc = "lower right")
ax.grid()
plt.show()
```

```python
X = np.arange(-5, 5, 0.1)
p = Polynomial(1, 0, -12, 0)
p_der = p.derivative
F = p(X)
F_derivative = p_der(X)

fig, ax = plt.subplots()
xpoint = [-2, 0, 2]
ypoint = [p(-2), p(0), p(2)]
ax.plot(X, F, label = f"$ {p}")
ax.plot(X, F_derivative, label = f"$ {p_der} $")
ax.annotate("local maximum", 
            (-2, p(-2)), 
            xytext = (-2+1,p(-2)+30), 
             arrowprops=dict(facecolor = "yellow", edgecolor = 'black', shrink = 0.05))
plt.annotate("inflection point",
             (0,p(0)),
              xytext=(0, p(0)+30), ##  
              arrowprops=dict(facecolor='yellow', ##   
                             edgecolor='black', ##   
                             fill=True, ##  (fill white  )
                             shrink=0.15, ##      ? 0  1   
                             alpha=0.8, ## , line, fill     . 
                             )
            )
ax.annotate("local minimum", 
            (2, p(2)),
            xytext = (2,p(2)+30), 
            arrowprops=dict(facecolor = "yellow", edgecolor = 'black', shrink = 0.05))
ax.legend(loc = "lower right")
ax.plot(xpoint, ypoint, 'ro')
ax.grid()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2) + np.cos(x)
derivative = 2 * x * np.cos(x**2) - np.sin(x)

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, figsize = (11,8))
fig.subplots_adjust(hspace = 0.8, wspace = 0)
ax1.plot(x, y, 'r--')
ax1.set_xlabel("value")
ax1.set_ylabel("counter")
ax1.set_title("sin + cos")
ax2.plot(x, derivative, 'c')
ax2.set_xlabel("value")
ax2.set_ylabel("counter")
ax2.set_title("derivative")
plt.show()
```
